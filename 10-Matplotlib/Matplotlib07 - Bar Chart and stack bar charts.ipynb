```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
company = ['Amazon','Google','Facebook','Apple']

Revenue = [120,90,70,75]
```

```python
plt.bar(company,Revenue,color = 'skyblue',edgecolor = 'black')
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
plt.bar(company,Revenue,color = 'skyblue',edgecolor = 'black')
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.tight_layout()
for i,j in zip(company,Revenue):
    plt.annotate(j,(i,j),color='red',size = 12)
plt.show()
```

```python
#lets to improve the visibility of data lables

plt.bar(company,Revenue,color = 'skyblue',edgecolor = 'black')
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.tight_layout()
for i,j in zip(company,Revenue):
    plt.annotate(j,(i,j),color='red',size = 12,ha='center',va='top')  #ha could be =left,right,center
                                                                      #va = top,bottom
plt.show()
```

```python
#Note

#ha = horizontal align

#va = vertical align
```

# All the bar charts above are stacked bar

# Here comes the question how we can see all these bars next to each other with respective to each company

lets try to break this

```python
#here we have get the help of numpy module

Index = np.arange(len(company))
Index
```

```python

plt.figure(figsize = (10,4))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

plt.bar(Index,Revenue)

plt.tight_layout()
#plt.legend()
plt.show()
```

```python
#Here on x axis we are getting weird numbers not company name , how we can get that?
```

```python
plt.figure(figsize = (10,4))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

plt.bar(Index,Revenue,label ='Revenue')
plt.xticks(Index,company)                #Here its mapping company name to xaxis index
plt.tight_layout()
#plt.legend()
plt.show()
```

```python
#Now second setp is adding other bars next to each other

# so place the bars next to each other, major role is of widht size

#Rember we have add the same width of first bar for the second for bar index location so that it can place
#exact after that

#so its like   first bar index at 0+width = position of next bar
```

```python
plt.figure(figsize = (10,4))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

plt.bar(Index,Revenue,label ='Revenue',width = 0.4)
plt.bar(Index+0.4,profit,label ='Profit',width = 0.4)
plt.xticks(Index,company)                #Here its mapping company name to xaxis index
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#In above chart you can see that center location for each bars still align to first bar(revenue bar) only

#lets fix this
```

```python
#we have to adjust the xticks width accordingly 

plt.figure(figsize = (12,4))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

plt.bar(Index,Revenue,label ='Revenue',width = 0.4)
plt.bar(Index+0.4,profit,label ='Profit',width = 0.4)
plt.xticks(Index+(0.4/2),company)                #Here we added the half of width of first bar to fix location
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#lets add third bar

w = 0.2
plt.figure(figsize = (12,4))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

bar1 = np.arange(len(company))
bar2 =[i+w for i in bar1]   # adding width to first index location so that in can place next to first bar
bar3 =[i+w for i in bar2]   # adding width to second bar 
plt.bar(bar1,Revenue,label ='Revenue',width = w)
plt.bar(bar2,profit,label ='Profit',width = w)
plt.bar(bar3,cost,label ='cost',width = w)
plt.xticks(Index+w,company)                #Here we added the width to adjust xtickt center location
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#Now we are adding data lable for each bar

w = 0.2
plt.figure(figsize = (12,5))
plt.title('Company Revenue')
plt.xlabel('Company')
plt.ylabel('Revenue')

bar1 = np.arange(len(company))
bar2 =[i+w for i in bar1]   # adding width to first index location so that in can place next to first bar
bar3 =[i+w for i in bar2]   # adding width to second bar 

plt.bar(bar1,Revenue,label ='Revenue',width = w)
for i,j in zip(bar1,Revenue):
    plt.annotate(j,(i,j),va='bottom',ha='center')
    
plt.bar(bar2,profit,label ='Profit',width = w)
for i,j in zip(bar2,profit):
    plt.annotate(j,(i,j),va='bottom',ha='center')
    
plt.bar(bar3,cost,label ='cost',width = w)
for i,j in zip(bar3,cost):
    plt.annotate(j,(i,j),va='bottom',ha='center')
plt.xticks(Index+w,company)                #Here we added the half of width of first bar to fix location
plt.tight_layout()
plt.legend()
plt.show()
```
