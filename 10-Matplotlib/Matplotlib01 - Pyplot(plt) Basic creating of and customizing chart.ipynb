```python
import matplotlib.pyplot as plt

#doc :https://matplotlib.org/tutorials/introductory/pyplot.html
```

```python
#lets create median salary vs age char

age = [25,26,27,28,29,30]
Median_salary_gp1 = [10000,30000,25000, 22000,100000,70000]
```

```python
plt.plot(age,Median_salary_gp1)
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.show()
```

```python
#Now suppose we want to compare this salary with same age group for different domain.
#how we can do that

Median_salary_gp2 = [10000,30000,50000,70000,90000,111000]

#we have to write that code again
plt.plot(age,Median_salary_gp1)

plt.plot(age,Median_salary_gp2) #adding for next line reference
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.show()

```

```python
#Now we can see the tow lines for different group under same age but we can't see properly 
#which line is denoting which group

# lets fix this

plt.plot(age,Median_salary_gp1,label='group1') #we have provided the label to this plot

plt.plot(age,Median_salary_gp2,label='group2') #provided lable for 2nd group plot
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.show()
```

```python
#adding color of our own choice


plt.plot(age,Median_salary_gp1,label='group1',color = 'red') #we have provided the label to this plot

plt.plot(age,Median_salary_gp2,label='group2',color='blue') #provided lable for 2nd group plot
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.show()
```

```python
#Changing line style
#The optional parameter *fmt* is a convenient way for defining basic formatting like 
# color, marker and linestyle.
plt.plot(age,Median_salary_gp1,label='group1',color = 'red',linestyle ='--') 

plt.plot(age,Median_salary_gp2,label='group2',color='blue',linestyle =':') 
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.show()
```

```python
#marker

#Changing line style
#The optional parameter *fmt* is a convenient way for defining basic formatting like 
# color, marker and linestyle.
plt.plot(age,Median_salary_gp1,label='group1',color = 'red',linestyle ='--',marker ='o') 

plt.plot(age,Median_salary_gp2,label='group2',color='blue',linestyle =':',marker = '+') 
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.show()
```

```python
#if we feel any line is bit thin or thicker we can change the width accordingly

plt.plot(age,Median_salary_gp1,label='group1',color = 'red',linestyle ='--',marker ='o') 

plt.plot(age,Median_salary_gp2,label='group2',color='blue',linestyle =':',marker = '+',linewidth=2) 
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.show()
```

```python
#to add the grid , to see better viual when the line cross particular point

#if we feel any line is bit thin or thicker we can change the width accordingly

plt.plot(age,Median_salary_gp1,label='group1',color = 'red',linestyle ='--',marker ='o') 

plt.plot(age,Median_salary_gp2,label='group2',color='blue',linestyle =':',marker = '+',linewidth=2) 
plt.ylabel('Median_salary')
plt.xlabel('age')
plt.title('Median Salary VS Ages')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#If we want to add the datalabel point in chart

age = [25,26,27,28,29,30]
Median_salary_gp1 = [10000,30000,25000, 22000,100000,70000]

with plt.xkcd():  # comic style so that it will apply to inner statement only not all
    
    plt.plot(age,Median_salary_gp1,label='Python developer')
    plt.ylabel('Median Salary')
    plt.xlabel('Age')
    plt.title('Median Salary vs Ages')
    plt.legend()
    plt.grid(True)

    for i,j in zip(age,Median_salary_gp1):
        plt.annotate(j,(i,j))
    plt.show()
```
