# What is scatter plot:

Also called: scatter plot, X-Y graph

The scatter diagram graphs pairs of numerical data, with one variable on each axis, to look for a relationship between them. If the variables are correlated, the points will fall along a line or curve. The better the correlation, the tighter the points will hug the line. 

# When to use:

When to Use a Scatter Diagram

    When you have paired numerical data
    When your dependent variable may have multiple values for each value of your independent variable
    When trying to determine whether the two variables are related or not and how they related, negatively or        positively.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#lets take two pair of variables with random numbers, just to demonstrate the functionality of scatter plot

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

```

```python
plt.scatter(x,y,label= 'X VS Y')

plt.show()
```

```python
plt.scatter(x,y,label= 'X VS Y')

plt.title('Relation b/w x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()

plt.show()
```

```python
plt.scatter(x,y,label= 'X VS Y',color = 'red',s = 100, alpha = 0.7)

plt.title('Relation b/w x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
plt.scatter(x,y,label= 'X VS Y',c = 'red',s = y, alpha = 0.7,edgecolor ='black' )

plt.title('Relation b/w x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
# lets try to do this with some realtime data

df = pd.read_csv('http://bit.ly/drinksbycountry')
df.head()
```

```python
#lets try to check the relation between beer servings and total liter of pure alcohol in it


Serving = df.beer_servings

Alcohol = df.total_litres_of_pure_alcohol
```

```python
plt.scatter(Serving,Alcohol,s = Serving, alpha = 0.6 )

plt.title('Beer Serving Vs pure Alcohol in it')
plt.xlabel('Beer Serving')
plt.ylabel('Pure Alchol in Liter')
# plt.legend()
plt.tight_layout()

plt.show()
```

```python
plt.scatter(Serving,Alcohol,s = Serving, alpha = 0.6,edgecolor='pink' )

plt.title('Beer Serving Vs pure Alcohol in it')
plt.xlabel('Beer Serving')
plt.ylabel('Pure Alchol in Liter')
# plt.legend()
plt.tight_layout()

plt.show()
```

```python
#There is positve relation between beer serving and pure alcohol in it.In some cased with less beer serving,
# pure alcohol still high

```

```python
plt.figure(figsize= (20,7))
plt.scatter(Serving,Alcohol,s = Serving, alpha = 0.6,edgecolor='pink',c=Alcohol )

plt.title('Beer Serving Vs pure Alcohol in it')
plt.xlabel('Beer Serving')
plt.ylabel('Pure Alchol in Liter')

cbar = plt.colorbar()
cbar.set_label('Increase Alcohol rate')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize= (20,7))
plt.scatter(Serving,Alcohol,s=150, alpha = 0.6,edgecolor='pink',c=Alcohol )

plt.title('Beer Serving Vs pure Alcohol in it')
plt.xlabel('Beer Serving')
plt.ylabel('Pure Alchol in Liter')

cbar = plt.colorbar()
cbar.set_label('Increase Alcohol rate')
plt.tight_layout()

plt.show()
```
