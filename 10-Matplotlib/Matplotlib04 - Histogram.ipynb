```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#let suppose we had serveyed ages for bunch of pepole 

age = [10,13,13,14,16,20,20,20, 33, 23, 25, 26, 35, 25, 25, 30, 37, 38, 45, 50 , 55]
```

```python
plt.hist(age) #calling histogram 
plt.show()   # to show graph
```

```python
#but above one doesn't looks pretty

plt.hist(age, label = 'Ages',color= 'skyblue',edgecolor='black')
plt.title('Respondant Ages')
plt.xlabel('Ages')
plt.ylabel('No. of observations')
plt.legend
plt.show()
```

```python
#these are creating random bins ,we can create our own too

plt.hist(age, label = 'Ages',color= 'skyblue',edgecolor='black',range= (10,60))
plt.title('Respondant Ages')
plt.xlabel('Ages')
plt.ylabel('No. of observations')
plt.legend
plt.show()
```

```python
#Above we gives range , but we can decide our bins too
plt.hist(age, label = 'Ages',color= 'skyblue',edgecolor='black',bins= [10,20,30,40,50,60])
plt.title('Respondant Ages')
plt.xlabel('Ages')
plt.ylabel('No. of observations')
plt.legend
plt.show()
```

```python
#Above we gives 6 bins with width/rang of 10
```

```python
#we want to add some additional insights like median reference line, we can do that also
```

```python
df = pd.DataFrame(age,columns =['Age'])
df
```

```python
median_age = df.Age.median()
```

```python
median_age
```

```python
plt.hist(age, label = 'Ages',color= 'skyblue',edgecolor='black',bins= [10,20,30,40,50,60])
plt.title('Respondant Ages')
plt.xlabel('Ages')
plt.ylabel('No. of observations')
plt.axvline(median_age,color= 'red',label= 'Median Age',linewidth = 1)
plt.legend()
plt.show()
```

```python
import numpy as np

ar = np.array(age)
ar
```

```python
plt.hist(age, label = 'Ages',color= 'skyblue',edgecolor='black',bins= [10,20,30,40,50,60])
plt.title('Respondant Ages')
plt.xlabel('Ages')
plt.ylabel('No. of observations')
plt.axvline(ar.mean(),color= 'red',label= 'Mean Age',linewidth = 1) # mean age
plt.legend()
plt.show()
```
