```python
import matplotlib.pyplot as plt
```

```python
slice = [60,40]

plt.pie(slice)
plt.title('My Pie Chart')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.pie(slice,labels= slice)
plt.title('My Pie Chart')
plt.show()
```

```python
slice1 = [60,40,30,20]
#plt.style.use('fivethirtyeight')
plt.pie(slice1,labels= slice1,wedgeprops ={'edgecolor':'black'})
plt.title('My Pie Chart')
plt.show()
```

```python
#Note : more than 5 to 10 items pie chart usally look very crowed so in this case pie chart will not be recommended
```

```python
slice1 = [60,40,30,20]
explode = [0,0,0.1,0] # we want to hight light any specific portion out of the whole pie chart,0.1 means 10% away 
                        #radius
plt.pie(slice1,labels= slice1,wedgeprops ={'edgecolor':'black'},explode = explode)
plt.title('My Pie Chart')
plt.show()
```

```python
slice1 = [60,40,30,20]
explode = [0,0,0.1,0] # we want to hight light any specific portion out of the whole pie chart,0.1 means 10% away 
                        #radius
plt.figure(figsize = (7,7))
plt.pie(slice1,labels= slice1,wedgeprops ={'edgecolor':'black'},explode = explode,shadow= True)
# Pie chart will give appearance of 3D
plt.title('My Pie Chart')
plt.show()
```

```python
#we want to show the percentage of each pie

slice1 = [60,40,30,20]
explode = [0,0,0.1,0] # we want to hight light any specific portion out of the whole pie chart,0.1 means 10% away 
                        #radius
plt.figure(figsize = (7,7))
plt.pie(slice1,labels= slice1,wedgeprops ={'edgecolor':'black'},explode = explode,shadow= True,autopct= '%.2f%%')
# Pie chart will give appearance of 3D
plt.title('My Pie Chart')
plt.show()
```
