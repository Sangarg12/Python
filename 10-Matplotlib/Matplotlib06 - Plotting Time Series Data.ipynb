```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime 
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates,y)

plt.title('Obvervations Vs time')
plt.xlabel('Date')
plt.ylabel('Events')
plt.show()
```

```python
plt.figure(figsize= (15,5))
plt.plot_date(dates,y)

plt.title('Obvervations Vs time')
plt.xlabel('Date')
plt.ylabel('Events')
plt.show()
```

```python
plt.figure(figsize= (15,5))
plt.plot_date(dates,y,linestyle= '--')

plt.title('Obvervations Vs time')
plt.xlabel('Date')
plt.ylabel('Events')
plt.tight_layout()
plt.show()
```

```python
#if suppose dates will show too conjusted then we fix this 
plt.figure(figsize= (15,5))
plt.plot_date(dates,y,linestyle= '--')
plt.gcf().autofmt_xdate()           # to auto fit the x axis date aligment, gcf(get the current figure)
plt.title('Obvervations Vs time')
plt.xlabel('Date')
plt.ylabel('Events')
plt.tight_layout()
plt.show()
```
