```python
# if we want all the visual on same figure we can use old method like

plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')
plt.plot(Age,javascript_dev,label ='javascript developer')
plt.plot(Age,All_dev,label ='All developer')

plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Now our requirement is we want the developer visual on different row and the python and jave dev in different row



fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1)
#(Here we are extracting the fig = number of figure then ax = axis of that figure)

ax1.plot(Age,All_dev,label ='All developer')
ax1.legend()

ax1.set_ylabel('Median Salary')

ax2.plot(Age,python_dev,label ='Python developer')
ax2.plot(Age,javascript_dev,label ='javascript developer')
ax2.legend()
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1,sharex=True)  #share x axis for both axis
#(Here we are extracting the fig = number of figure then ax = axis of that figure)

ax1.plot(Age,All_dev,label ='All developer',)
ax1.legend()

ax1.set_ylabel('Median Salary')

ax2.plot(Age,python_dev,label ='Python developer')
ax2.plot(Age,javascript_dev,label ='javascript developer')
ax2.legend()
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary')

plt.tight_layout()
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows = 2, ncols = 1,sharex=True,figsize=(12,5))  #share x axis for both axis
# to change the figure of subplots

ax1.bar(Age,All_dev,label ='All developer',)
ax1.legend()

ax1.set_ylabel('Median Salary')

ax2.plot(Age,python_dev,label ='Python developer')
ax2.plot(Age,javascript_dev,label ='javascript developer')
ax2.legend()
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary')

plt.tight_layout()
plt.show()
```

```python
#Now lets represent the both visuals next to each other 

fig, (ax1,ax2) = plt.subplots(nrows = 1, ncols = 2,sharey=True,figsize=(12,5))  #share x axis for both axis
# to change the figure of subplots

ax1.bar(Age,All_dev,label ='All developer',)
ax1.legend()
ax1.set_xlabel('Age')
ax1.set_ylabel('Median Salary')

ax2.plot(Age,python_dev,label ='Python developer')
ax2.plot(Age,javascript_dev,label ='javascript developer')
ax2.legend()
ax2.set_xlabel('Age')
#ax2.set_ylabel('Median Salary')

plt.tight_layout()
plt.show()
```
