# xlim & ylim

```python
plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')

left, right = plt.xlim()                             # to check current x axis intital and final limit
print(left,right)
plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#lets change the x axis left to right value from 10 to 60

plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')

                
left,right =plt.ylim()
print(left,right)

plt.xlim(10,60)         # to change the x axis limit  
plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
plt.show()
```

# Adjusting location of legend

```python
plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')
plt.plot(Age,javascript_dev,label = 'Javascript deveploper')

plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#deafullt legend location is best( to avoid minimus overlapping with others lines etc.)

# but if we want to change the location of legend position we can do that
```

```python
plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')
plt.plot(Age,javascript_dev,label = 'Javascript deveploper')

plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend(loc = 'lower right') # there is 9 to 10 positions which we can specifiy here with loc.
plt.tight_layout()
plt.show()
```

```python
# we can distribute the logends in rows and columns as per our use case or visual type

plt.figure(figsize = (12,5))
plt.plot(Age,python_dev,label ='Python developer')
plt.plot(Age,javascript_dev,label = 'Javascript deveploper')

plt.title('Developers Median salary veres Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.legend(loc = 'lower right',ncol=2) # there is 9 to 10 positions which we can specifiy here with loc.
plt.tight_layout()
plt.show()
```

# legend outside of the graph & legend title
