```python
#Lets create one dataframe to check how to handle missing value
import numpy as np
people = {'first_name':['Sandy','Mandy','Randy',np.nan,None,'NA'],
          'last_name' :['A','B','C',np.nan,np.nan,'missing'],
          'email' : ['a@gmail.com','b@gmail.com','c@gmail.com',None,np.nan,np.nan],
           'age'  : ['10','20','30',None,None,'missing']}
```

```python
pep = pd.DataFrame(people)
```

```python
pep
```

```python
# This data frame has bunch of missing value and custom missing value

pep.dtypes # it shows type of each columns
```

```python
pep.isna()  # to check which type it consider as missing or na value
```

```python
pep.isnull()
```

```python
#if we want to drop all null value in our dataset, but it is not ideal thing to do with missig value

pep.dropna()
```

```python
pep.dropna(axis=0,how='all') # in this case untill whole row won't null , it won't drop
```

```python
# this same thing we can along column

pep.dropna(axis=1,how='all') # axis = 0,1 means row, column respectively
```

```python
pep.dropna(axis=1,how='any') # if any null value drop in column
```

```python
pep # we can target specific column or row too w.r.t axis
```

```python
pep.dropna(axis=0,how='any',subset=['email']) # if any null value in email then remove that row
```

```python
# we can target multiple columns too

pep.dropna(axis=0,how='all',subset=['email','last_name']) #if both missing together than remove row
```

```python
#how to remove custom missing values

pep.replace(['NA','missing'],np.nan,inplace=True)
```

```python
pep
```

```python
pep.dropna(axis=0,how='any',subset=['email','last_name'],inplace=True)
```

```python
pep # you can see now we cleaned our data
```

```python
#now lets find average of age

pep.dtypes
```

```python
pep['age'].mean()
```

```python
#it shows wrong mean bcoz data type of age is not in int or float

#lets change it

pep['age']=pep['age'].astype(float)
```

```python
pep['age'].mean() # now we are getting correct average for age 
```
