# Making changes to the data

```python
#Adding new column with the help of existing columns

df.head(2)

df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']

df.head(3)
```

```python
df  # now you can see this new column has been added to the dataframe
```

```python
#droping of column

df=df.drop(columns=['Total'])
df.head(3)
```

```python
#Adding new column in fency way

df['Total']=df.iloc[:,4:10].sum(axis=1)  # axis means sum along column
df.head(3)
```

```python
#Rearraning columns

cols = list(df.columns.values)
```

```python
cols
```

```python
df=df[cols[0:4]+cols[-1:]+cols[4:12]]
df.head(5)
```

```python
df  # now we can see that total columns reference is changed
```

```python
df.set_index('#',inplace= True)  # reseting index column
```

```python
df
```

```python
df.rename(index={'#':'Index'},inplace= True)
```

```python
df.index.names = ['Sr.No.']
```

```python
df
```

```python
#now lets filter out with search of specific expression

import re
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex = True)]
# all the names start with pi following a to z with any repetition of match, flags= re.I means ingnore capitlizatioin
```

```python
# if we remove '^' then specified condition could be exist at any location in string
df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex = True)]
```
