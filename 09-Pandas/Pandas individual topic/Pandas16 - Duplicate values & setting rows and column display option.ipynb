```python
import pandas as pd
```

```python
movie = pd.read_csv('http://bit.ly/imdbratings')

movie.head()
```

```python
#lets check how many duplicated rows are there if any duplicated rows exist

movie.duplicated().sum()
```

```python
#there is 0 duplicat rows

# letss check specific column

movie['genre'].duplicated().sum()
```

```python
movie.shape
```

```python
#so out of 979, 963 entries in genre gets repeated , actually these are not duplicate as these are category
#but this is giving us idea how many entries gets repeated within same column keeping the first record aside.
```

```python
#how we can drop duplicate values

movie.drop_duplicates()
```

```python
movie['genre'].drop_duplicates().value_counts().sum()
```
