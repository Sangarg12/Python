```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
#method first

filter1= movies['duration']>=200
```

```python
movies[filter1]  # rows will appear with >=200 duration only
```

```python
#we can use condition directly also, which is nothing act like boolean

movies[movies.duration>=200]
```

```python
#and we want to see any see any specific column with that filtered condition

movies[movies.duration>=200]['title']
```

```python
#we do have best method than above one which we going to discuss below
```

# How do I apply multiple filter criteria to a pandas DataFrame?

```python
# simply we can use '&' = 'and' or '|'= 'OR' condition
```

```python
movies[(movies['duration']>=200) & (movies['genre']=='Action')] #and condition

#Note: we need to use '&' or pipe('|') , here 'and' and 'or' will not work
```

```python
movies[(movies['duration']>=200) | (movies['genre']=='Action')] # or condition
```

```python
#another method which is equivalent to 'OR' condition

movies[movies['genre'].isin(['Action','Crime'])]
```

```python
#Best method to check columns with filter condition

movies.loc[(movies['duration']>200)&(movies['genre']=='Drama'),'star_rating':'genre']
```

```python
#not in filter

movies.loc[~(movies['duration']>=100)& ~(movies['genre'].isin(['Drama','Crime','Action','Comedy'])),'star_rating':'duration']
```

```python
#kind of regular expression filter
movies[movies['title'].str.contains('Man',na=False)]
```
