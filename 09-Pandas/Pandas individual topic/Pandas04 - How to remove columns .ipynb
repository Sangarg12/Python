# How to remove columns from pandas dataframe

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')

ufo.head(3)
```

```python
ufo1 = ufo.copy()
```

```python
#we can drop single column or multiple column by passing list of it
ufo1.drop(['Colors Reported','State'],axis=1,inplace= True)
```

```python
ufo1  # ta da columns has been droped across columns axis
```

```python
ufo # we made changes to the copied data frame only, this is best practice to reservse the original data as it is
    # so that if some changes made by us inplace mistankely we can go back to orignal dataframe
```

```python
#Similary we can drop rows

ufo1.drop([0,1,2],axis = 0 ,inplace = True)

ufo1.head(10)
```

```python
#We can see above that first 3 rows has been eliminated permanently from dataframe
```
