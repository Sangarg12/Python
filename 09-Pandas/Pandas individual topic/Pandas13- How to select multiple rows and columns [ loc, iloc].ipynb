```python
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/uforeports')

df.head()
```

```python
#loc method , loc use to select rows and columns by labels


df.loc[0,:] # first row and all columns
```

```python
df.loc[[0,1,2],:] # first 3 rows with all columns
```

```python
#above result with indexing method

df.loc[0:2,:]
```

```python
df.loc[:,['City','State']] # all rows and selected columns
```

```python
df.loc[:,'City':'State'] # all rows with column range
```

```python
#filtered sepecfic column with selected columns

df.loc[df['City']=='Holyoke',['City','State']]
```

```python
# Same above thing we can do :

df[df['City']=='Holyoke'][['City','Shape Reported','State']]
```

# iloc


```python
#iloc is selecting columns and rows by index location

df.iloc[:,0:2] # all rows , first two columns
```

```python
df.iloc[5:11,:] # range of rows and all columns
```

```python
df.iloc[5:11,'City'] # we can't put lables in iloc as this call by indexing only
```
