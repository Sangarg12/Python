# How do I sort values in pandas dataframe

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.sort_values(['title']) # sorting single column
```

```python
#sorting multiple columns 

movies.sort_values(['genre','duration'])
```

```python
#sorting multiple columns with ascending and descending format

movies.sort_values(['star_rating','genre','duration'],ascending=[False,True,False],inplace=True)
```
